Gráficos Básicos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos de Ejemplo
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
#Gráficos de líneas

plt.figure(figsize=(5,2))
plt.plot(x,y)
plt.title('Gráfico de líenas')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Gráfico de Barras

categorias = ['A','B','C','D']
valores = [3,7,2,5]

plt.figure(figsize=(5,2))
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
#Grafico de Disperción

x= np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10,4))
plt.scatter(x,y)
plt.title('Gráfico de Dispercion')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Histograma

datos=np.random.rand(1000)

plt.figure(figsize=(10,4))
plt.hist(datos,bins=30)
plt.title('Histograma')
plt.xlabel('valores')
plt.ylabel('frecuencia')
plt.show()
```
